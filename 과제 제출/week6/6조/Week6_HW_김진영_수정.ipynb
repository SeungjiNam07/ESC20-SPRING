```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA, FactorAnalysis
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('wine.csv',names=['class','Alcohol','Malic acid','Ash','Alcalinity of ash','Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline'])
df.head()
```

# PCA 및 FA 적용 전/후 분류모델 성능 비교 (Logistic Regression 모델 이용)

## PCA 및 FA 적용 전

```python
df.head()
```

```python
features=['Alcohol','Malic acid','Ash','Alcalinity of ash','Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines', 'Proline']
y=df['class'].values
X=df[features].values #design matrix
```

```python
# train-test 분류 (8:2): Scikit learn의 train_test_split 함수 이용
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

```python
# 분류모델: Logistic Regression
model = LogisticRegression(solver='sag', max_iter=100000)
```

```python
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
print(accuracy_score(y_test,pred))
```

=> PCA 및 FA 적용 전 분류모델 성능: 97% 정확도

## PCA 적용 후

```python
X=df.loc[:, features].values
y = df.loc[:, ['class']].values

#Scaling
X_sca = StandardScaler().fit_transform(X) 

# https://subinium.github.io/MLwithPython-3-3/
# StandardScaler(): 각 feature 평균0, 분산1이 되도록 변환
# fit(): train data로 모델 파라미터를 학습시킴
# transform(): 학습된 파라미터로 데이터 변환시킴
```

```python
pd.DataFrame(data=X_sca, columns=features).head()
```

```python
# PCA 적용
pca = PCA(n_components=2) # PCA 선언 : 2차원으로 차원 축소 => 제1~2 주성분, 총 2개의 주성분 반환
X_pca = pca.fit_transform(X_sca) # PCA 적합

df_pca = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])
df_pca = pd.concat([df[['class']], df_pca], axis=1)
df_pca.head()
```

```python
y_pca=df_pca['class'].values

# train-test 분류 (8:2)
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y_pca, test_size=0.2, random_state=7)
```

```python
# PCA 적용 전과 같은 분류모델 사용
model2 = LogisticRegression(solver='sag', max_iter=100000)
```

```python
model2.fit(X_train_pca,y_train_pca)
```

```python
pred_pca=model2.predict(X_test_pca)
print(accuracy_score(y_test_pca,pred_pca))
```

=> PCA 적용 후 분류모델 성능: 약 97% 정확도

- 적용 전과 유사
- 근데 random_state를 다르게 해서 여러번 시도해본 결과 pca로 했을 때 더 낮게 나오는 경우도 있고 높은 경우도 있고 계속 다름...
- 결론) PCA 여부가 분류모델의 성능에 영향을 미치지 않는 것 같다.

참고) pipeline 써서 pca + 분류 한번에 해도 됩니다,,, 전 익숙치않아서,,, 아직,,ㅎㅎ;;

## FA 적용 후

```python
X = df.loc[:, features].values
fa = FactorAnalysis(n_components=13)
X_fa = fa.fit_transform(X)
```

```python
df_fa = pd.DataFrame(data=X_fa, columns=['Factor1', 'Factor2', 'Factor3', 'Factor4', 'Factor5', 'Factor6', 'Factor7', 'Factor8', 'Factor9', 'Factor10', 'Factor11','Factor12', 'Factor13'])
df_fa = pd.concat([df[['class']],df_fa], axis=1)
df_fa.head()
```

Factor 1~7까지 유의미

```python
y_fa=df_fa['class'].values
X_train_fa, X_test_fa, y_train_fa, y_test_fa = train_test_split(X_fa, y_fa, test_size=0.2, random_state=7)
```

```python
# FA 적용 전과 같은 분류모델
model3 = LogisticRegression(solver='sag', max_iter=100000)
```

```python
model3.fit(X_train_fa,y_train_fa)
```

```python
pred_fa=model3.predict(X_test_fa)
print(accuracy_score(y_test_fa,pred_fa))
```

=> PCA 적용 후 분류모델 성능: 100% 정확도

- 적용 전보다 다소 증가했다! (97->100%) 
- 근데 random_state를 다르게 해서 여러번 시도해본 결과 pca와 마찬가지로 결과는 계속 다르나 유사함
- 결론) PCA와 마찬가지로 FA 여부가 분류모델의 성능에 영향을 미치지 않는 것 같다...
