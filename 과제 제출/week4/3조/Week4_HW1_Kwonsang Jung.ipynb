## Load Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
```

```python
iris = datasets.load_iris()
X = iris.data[:,:2]
y = iris.target
```

## SVM(linear) with C = 1, gamma = 0.1

```python
C = 1.0
svc = svm.SVC(kernel='linear', C=1, gamma=0.1).fit(X, y)
```

```python
x_min, x_max = X[:,0].min() - 1, X[:,0].max() + 1
y_min, y_max = X[:,1].min() - 1, X[:,1].max() + 1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
```

```python
plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.8)

plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with linear kernel')
plt.show()
```

## SVM(rbf) with gamma = 0.1, 10, 100

```python
svc = svm.SVC(kernel='rbf', C=1, gamma=0.1).fit(X, y)
```

```python
gamma = [0.1, 10, 100]
for i in gamma:
    svc = svm.SVC(kernel='rbf', C=1, gamma=i).fit(X, y)
    plt.subplot(1, 1, 1)
    Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8)

    plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(), xx.max())
    plt.title('SVC with rbf kernel')
    plt.show()
```

## SVM with C = 1, 100, 1000

```python
C = [1, 100, 1000]
for i in C:
    svc = svm.SVC(kernel='rbf', C=i, gamma=0.1).fit(X, y)
    plt.subplot(1, 1, 1)
    Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8)

    plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(), xx.max())
    plt.title('SVC with rbf kernel')
    plt.show()
```
