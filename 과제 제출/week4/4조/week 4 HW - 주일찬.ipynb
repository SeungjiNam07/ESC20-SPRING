```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
```

```python
# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2] 
# we only take the first two features. We could avoid this ugly slicing by using a two-dim dataset
y = iris.target
```

```python
# we create an instance of SVM and fit out data. We do not scale our data since we want to plot the support vectors
C = 1.0 
# SVM regularization parameter
svc = svm.SVC(kernel='linear', C=1,gamma='auto').fit(X, y)
```

```python
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
 np.arange(y_min, y_max, h))
```

```python
plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with linear kernel')
plt.show()
```

## Imbalanced Classification Dataset

```python
from sklearn.datasets import make_classification
```

```python
# define dataset
# 10,000 examples with an approximate 1:100 minority to majority class ratio.(100개 & 9900개)
X, y = make_classification(n_samples=10000, n_features=2, n_redundant=0,
	n_clusters_per_class=1, weights=[0.99], flip_y=0, random_state=4)
```

```python
# summarize class distribution
from collections import Counter
counter = Counter(y)
print(counter)
```

데이터셋이 예상하는 개수대로 잘 분류되었다.

```python
# scatter plot of examples by class label
for label, _ in counter.items():
    row_ix = np.where(y == label)[0]
    plt.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))

plt.legend()
plt.show()
```

## Weighted SVM With Scikit-Learn

```python
from numpy import mean
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.svm import SVC
```

```python
# define model
model = SVC(gamma='scale', class_weight='balanced')
```

```python
# define evaluation procedure
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate model
scores = cross_val_score(model, X, y, scoring='roc_auc', cv=cv, n_jobs=-1)
# summarize performance
print('Mean ROC AUC: %.3f' % mean(scores))
```
