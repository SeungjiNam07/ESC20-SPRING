corr 이 좋아보이는 od280/od315 / flavanoid /Total phenols 를 선택하자.

```python
data = wine[['name','Total phenols', 'Flavanoids', 'OD280/OD315 of diluted wines']]
```

```python
X = data.iloc[:,1:]
Y = data['name']
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=30)

print(X_train.shape)
print(X_test.shape)
X_train.head()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis().fit(X_train, Y_train)
print('LDA Train', round(lda.score(X_train, Y_train),3))
print('LDA Test', round(lda.score(X_test, Y_test),3))
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
qda = QuadraticDiscriminantAnalysis().fit(X_train,Y_train)
print('QDA Train', round(qda.score(X_train, Y_train),3))
print('QDA Test', round(qda.score(X_test, Y_test),3))
```

```python
from sklearn.linear_model import LogisticRegression
Logi = LogisticRegression(solver='sag', max_iter=10000, multi_class='auto').fit(X_train, Y_train)
print('Logi Train', round(Logi.score(X_train, Y_train),3))
print('Logi Test', round(Logi.score(X_test, Y_test),3))
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import cross_val_score
for i in range(1,4):
    poly = PolynomialFeatures(degree=i, include_bias=False)
    X_poly = poly.fit_transform(X_train)
    log = LogisticRegression(solver='sag', max_iter=10000, multi_class='auto')
    kcv_score = cross_val_score(log, X_poly, Y_train.values.ravel(), cv=5) 
    print('kcv degree',i, ':', round(kcv_score.mean(),3))
```

```python
# TEST SCORE 를 비교해본 결과 LDA > QDA > LOGI
```
